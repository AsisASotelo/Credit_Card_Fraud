# Initial Data Exploration

The purpose of this notebook is to assess data quality of the data located in the "creditcard".csv

### Content Of Dataset

The data contains credit card transaction data from September 2013. It contains data that occured in the span of two days and contains 492 fraudulent entries and a total of 284,807 transactions. About .172 % of the data is fraudelant. This means that the dataset is **unbalanced**. Certain actions must be taken and specific models will be used keeping this in mind. Couple of things to note from others who have done models on this. 

* Do not test models on oversampled or undersampled data
* If you want to implement cross validation, oversample and undersample during cross validation
* Don't use accuracy score for anamoly detection us f1, precision/recall, or confusion matrix

A couple of things to note about the features:
* The data has had PCA performed on it from V1-V28 so we assume that it has been scaled
